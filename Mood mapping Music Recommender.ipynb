```python
!pip install ipywidgets
import gradio as gr
from textblob import TextBlob
from .autonotebook import tqdm as notebook_tqdm
```

```python
# Mood-to-genre mapping
mood_genre_map = {
    "happy": ["pop", "dance", "upbeat rock"],
    "sad": ["acoustic", "blues", "soft piano"],
    "angry": ["heavy metal", "hard rock", "rap"],
    "anxious": ["ambient", "classical", "chill-out"],
    "relaxed": ["jazz", "lo-fi", "soul"],
    "energetic": ["edm", "hip-hop", "pop rock"],
    "romantic": ["r&b", "soft pop", "love ballads"],
    "lonely": ["indie", "acoustic", "folk"],
    "neutral": ["ambient", "electronic", "instrumental"]
}
```

```python
# Sentiment analyzer
def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0.5:
        return "happy"
    elif 0 < polarity <= 0.5:
        return "relaxed"
    elif polarity == 0:
        return "neutral"
    elif -0.5 < polarity < 0:
        return "anxious"
    else:
        return "sad"
    
```

```python
# Recommender function
def mood_music_recommender(user_input):
    interpreted_mood = analyze_sentiment(user_input)
    genres = mood_genre_map.get(interpreted_mood, ["ambient"])
    return {
        "user_mood": user_input,
        "interpreted_mood": interpreted_mood,
        "recommended_genres": genres,
        "mapping_explanation": f"Using TextBlob polarity to detect '{interpreted_mood}', mapped based on mood-genre schema."
    }
```

```python
# Gradio interface
iface = gr.Interface(
    fn=mood_music_recommender,
    inputs=gr.Textbox(lines=2, placeholder="Describe your current mood..."),
    outputs="json",
    title="ðŸŽµ Mood-Based Music Recommender ",
    description="Enter your current feelings and get genre suggestions!"
)

iface.launch()
```
